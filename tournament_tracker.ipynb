```python
import csv
print("Welcome to TournamentTracker!")
print("Select how many places to create: ")
totalEntries = int(input())
print(f"This tournament is set for {totalEntries} participants")

entrantDict = {}
for i in range(1, totalEntries + 1):
    entrantDict[i] = None

MainMenu()
```

```python
def addEntrant():
    global entrantDict
    global totalEntries
    print(f"Select the position out of {totalEntries}")
    addSelection = int(input())
    if entrantDict[addSelection] == None:
        print("Enter the name of the participant: ")
        entrantName = input()
        entrantDict[addSelection] = entrantName
        print(f"Position {addSelection} is now held by {entrantName}")
    else:
        print("Invalid Entry, position already taken.")
```

```python
def delEntrant():
    global entrantDict
    global totalEntries
    print(f"Select a position out of {totalEntries}")
    delSelection = int(input())
    print(f"What is the name of entrant #{delSelection}?")
    delName = input()
    
    if entrantDict[delSelection] == delName:
        entrantDict[delSelection] = None
        print(f"Deletion successful! position {delSelection} is now open to be added.")
    else:
        print("Invalid Entry")
```

```python
def viewEntrants():
    global totalEntries
    global entrantDict
    print(f"Select a Position to view from out of {totalEntries}:")
    viewPosition = int(input())
    
    startingPosition = viewPosition - 5
    if startingPosition <= 0:
        startingPosition = 1
    endingPosition = viewPosition + 5
    if endingPosition > len(entrantDict):
        endingPosition = len(entrantDict)
    
    for i in range(startingPosition, endingPosition + 1):
        print(f"#{i}: {entrantDict[i]}")
```

```python
def saveCSV():
    print("Would you like to save your csv? Select Y/N")
    save = input()
    if save == 'Y':
        global entrantDict
        csvFileName = "TournamentPositions.csv"
        with open(csvFileName, "w") as file:
            writer = csv.writer(file)
            for key, value in entrantDict.items():
                writer.writerow([key, value])
        print('Save Successful.')
    else:
        print("Save Canceled.")
```

```python
def MainMenu():
    operating = False
    while operating != True:
        print("======================TournamentTracker======================\nSelect an option: \n        1. Add an Entrant\n        2. Delete Entrant\n        3. View Entrants\n        4. Save Changes\n        5. Exit Program")
        userInput = input()
    
        if userInput == '1':
            addEntrant()
        elif userInput =='2':
            delEntrant()
        elif userInput == '3':
            viewEntrants()
        elif userInput == '4':
            saveCSV()
        elif userInput == '5':
            print("Are you sure you want to quit? Any unsaved progress will be lost. Select Y/N")
            quitChoice  = input()
            if quitChoice == 'Y':
                print("Goodbye!")
                operating = True
            else:
                print('Exit Cancelled.')
        else:
            print("Invalid selection, please try again.")
```
